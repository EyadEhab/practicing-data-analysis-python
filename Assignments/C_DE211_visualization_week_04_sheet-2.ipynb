#Get Started with Data Visualization in Python

### Step 1: Import needed liberaries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Gather and Showcase the Dataset

you can fitch and download the dataset from this url https://https://www.kaggle.com/datasets/kkhandekar/cost-of-living-index-by-city-2022?resource=download

about the dataset: Cost of living index for 578 cities - 2022


```python
df = pd.read_csv('costofliving.csv')

df.head()
```

### Step 3: Understand the Data

show some information about the dataset

```python
df.info()

df.describe()
```

### Step 4: Clean and Organize the Data

As you can see, the names of the columns are very long. So, let’s rename them. We can use the Pandas rename() method to rename any index, column, or row:

```python
df.rename(columns={
    'Cost of Living Index': 'COL_Index',
    'Rent Index': 'Rent_Index',
    'Cost of Living Plus Rent Index': 'COL_Rent_Index',
    'Groceries Index': 'Groceries_Index',
    'Restaurant Price Index': 'Restaurant_Index',
    'Local Purchasing Power Index': 'Purchasing_Power_Index'
}, inplace=True)
```

#### Now that we’ve successfully renamed the columns, let’s take a look at the countries.


```python
df.head()
```

### Step 5: Data Visualization


One of the simplest ways to get started visualizing data is to use the Matplotlib package’s submodule pyplot (plt) to create a quick graph of your data.

For example, the following code should show the rent over cost of living displayed in a **scatter plot** graph:

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['COL_Index'], df['Rent_Index'])
plt.title('Cost of Living Index vs Rent Index')
plt.xlabel('Cost of Living Index')
plt.ylabel('Rent Index')
plt.show()
```

### Next, let’s use another package called Seaborn (sns) to visualize what rent and groceries look like.

Do you see a correlation here? Overall, as the price of rent increases, we see that the cost of groceries increases as well.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['Rent_Index'], y=df['Groceries_Index'])
plt.title('Rent Index vs Groceries Index')
plt.xlabel('Rent Index')
plt.ylabel('Groceries Index')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(df['City'], df['Rent_Index'])
plt.title('City vs Rent Index')
plt.xlabel('City')
plt.ylabel('Rent Index')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='City', y='COL_Index', data=df)
plt.title('Cost of Living Index by City')
plt.xlabel('City')
plt.ylabel('Cost of Living Index')
plt.xticks(rotation=90)
plt.show()
```
